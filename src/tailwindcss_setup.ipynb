## 1. Open your terminal (CMD or PowerShell)

Navigate to your project folder:

```bash
cd path/to/your/project
```

---

## 2. Install Tailwind CSS (CLI version)

Run the following command to install Tailwind and its CLI helper:

```bash
npm install tailwindcss @tailwindcss/cli
```

---

## 3. Import Tailwind into your CSS

Create or open your main CSS file (e.g. `src/input.css`) and add:

```css
@import "tailwindcss";
```

This line tells your project to use all Tailwind base, components, and utilities.

---

## 4. Build your Tailwind output file

Run the Tailwind CLI to scan your files and generate the CSS output:

```bash
npx @tailwindcss/cli -i ./src/input.css -o ./src/output.css --watch
```

> ðŸ’¡ The `--watch` flag automatically rebuilds your CSS whenever you save changes.

---

## 5. Link your generated CSS in HTML

In your `index.html`, make sure to include:

```html
<link rel="stylesheet" href="./src/output.css">
```
